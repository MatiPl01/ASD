> Proszę zaimplementować funkcję, która otrzymuje na wejściu posortowaną niemalejąco tablicę A o rozmiarze n oraz liczbę x i sprawdza, czy x występuje w A. Jeśli tak, to zwraca najmniejszy indeks, pod którym x występuje.

### Implementacja algorytmu

```python
def find_first(arr, val):
    i = 0
    j = len(arr) - 1
    while i <= j:
        center_idx = (i + j) // 2
        if val > arr[center_idx]:
            i = center_idx + 1
        else:
            j = center_idx - 1
    return i if i < len(arr) and arr[i] == val else None
```

##### Implementacja funkcji testującej prawidłowość algorytmu (nie jest częścią zadania)

```python
import random

def test_find(fn, *, reps=20):
    passed = 0
    for i in range(reps):
        values = sorted(random.randint(-10, 10) for _ in range(random.randint(5, 25)))
        searched = random.randint(-10, 10)
        result = fn(values, searched)
        try:
            expected = values.index(searched)
        except ValueError:
            expected = None
        is_correct = expected == result
        passed += is_correct
        print(f'#{i+1} Test (already passed: {passed}/{reps})')
        print('Values:', values)
        print('Searched value:', searched)
        print('Expected:', expected)
        print('Result:', result)
        print(f'Test {"passed" if is_correct else "failed"}')
        print()
    print(f'Total passed: {passed}/{reps}')
    print(f'An algorithm is {"correct" if passed == reps else "wrong"}')
```

###### Przykładowy test

```python
test_find(find_first, reps=100)
```

### (Dodatkowo) Implementacja algorytmu, który zwraca indeks ostatniej przekazanej do wyszukania wartości

```python
def find_last(arr, val):
    i = 0
    j = len(arr) - 1
    while i <= j:
        center_idx = (i + j) // 2
        if val < arr[center_idx]:
            j = center_idx - 1
        else:
            i = center_idx + 1
    return j if j < len(arr) and arr[j] == val else None
```

##### Implementacja funkcji testującej prawidłowość algorytmu (nie jest częścią zadania)

```python
import random

def test_find(fn, *, reps=20):
    passed = 0
    for i in range(reps):
        values = sorted(random.randint(-10, 10) for _ in range(random.randint(5, 25)))
        searched = random.randint(-10, 10)
        result = fn(values, searched)
        expected = None
        found = False
        for i in range(len(values)):
            if searched == values[i]:
                expected = i
                found = True
            elif found:
                break
        is_correct = expected == result
        print(f'#{i+1} Test (already passed: {passed}/{reps})')
        print('Values:', values)
        print('Searched value:', searched)
        print('Expected:', expected)
        print('Result:', result)
        print(f'Test {"passed" if is_correct else "failed"}')
        print()
        passed += is_correct
    print(f'Total passed: {passed}/{reps}')
    print(f'An algorithm is {"correct" if passed == reps else "wrong"}')
```

###### Przykładowy test

```python
test_find(find_last, reps=100)
```
