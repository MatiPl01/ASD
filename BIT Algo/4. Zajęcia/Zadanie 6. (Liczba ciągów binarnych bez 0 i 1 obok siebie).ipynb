#### Top-down

```python
def count_seq(n: 'length of a sequence'):
    cache = [0] * (n + 1)
    
    def count(i: 'length of a subsequence'):
        if i <= 0: return 1
        if not cache[i]:
            cache[i] = count(i - 1) + count(i - 2)
        return cache[i]
        
    return count(n)
```

###### Kilka testów

```python
for n in range(11):
    print(count_seq(n), end=' ')
```

#### Bottom-up

Jak liczenie kolejnych wyrazów ciągu Fibonacciego

```python
def count_seq(n: 'length of a sequence'):
    if n < 2: return n + 1
    cache = [0] * (n + 1)
    cache[0] = 1
    cache[1] = 2
    
    for i in range(2, n + 1):
        cache[i] = cache[i - 1] + cache[i - 2]
        
    return cache[-1]
```

###### Kilka testów

```python
for n in range(11):
    print(count_seq(n), end=' ')
```

#### Zwykła iteracja (tu najszybsza i złożoność pamięciowa to $ O(1) $)

Jak liczenie kolejnych wyrazów ciągu Fibonacciego

```python
def count_seq(n: 'length of a sequence'):
    f1 = 1
    f2 = 2
    for i in range(n):
        f1, f2 = f2, f1 + f2
    return f1
```

###### Kilka testów

```python
for n in range(11):
    print(count_seq(n), end=' ')
```
