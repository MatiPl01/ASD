#### Top-down
##### Recursive with memoization

```python
def cut_rod(P: 'profits', n: 'length of a rod to cut'):
    cache = [0] * (n + 1)
    cache[1] = P[1]
    
    def cut(m: 'length of a current part of a rod to cut'):
        if m == 0: return 0
        if cache[m]: return cache[m]
        
        best = 0
        for i in range(1, m + 1):
            best = max(best, P[i] + cut(m - i))
        cache[m] = best
        return best
       
#     return cut(n)
    cut(n)
    print(cache)
    
    return cache[-1]
```

###### Kilka testów

```python
P = [0, 1, 3, 6, 5, 3]  # Zaczynamy od wartości zero, bo dla pręta o długości 0 nie mamy zysku
n = 5

print(cut_rod(P, n))
```

#### Bottom-up
##### iterative

```python
def cut_rod(P: 'profits', n: 'length of a rod to cut'):
    cache = [0] * (n + 1)
    cache[1] = P[1]
    
    for i in range(1, n + 1):
        for j in range(i):
            cache[i] = max(cache[i], cache[j] + P[i - j])
    
    print(cache)
    
    return cache[n]
```

###### Kilka testów

```python
P = [0, 1, 3, 6, 5, 3]  # Zaczynamy od wartości zero, bo dla pręta o długości 0 nie mamy zysku
n = 5

print(cut_rod(P, n))
```
