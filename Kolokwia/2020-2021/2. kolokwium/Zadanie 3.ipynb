### Omówienie algorytmu

W pierwszej kolejności przemapowuję tablicę wejściową na tablicę krotek
postaci (indeks początku plamy, ilość paliwa). Następnie, korzystając z kolejki
priorytetowej, w każdym kroku wybieram plamę, na której mogę zatankować najwięcej
w dystansie, na jaki jestem w stanie dojechać. Zauważmy, że w ten sposób algorytm
niejako może sprawdzać plamy, które już minęliśmy, ale nie jest to problem, ponieważ
wybór najlepszego rozwiązania w danym momencie gwarantuje nam największy zasięg.
To cofnięcie się utożsamiam z przypadkiem, gdy po dojechaniu do najbardziej opłacalnej
plamy, okzazuje się, że dalej już nie opłaca sie tankować, lub konieczne było tankowanie
wcześniej. Musimy więc zawsze sprawdzać wszystkie plamy, jakie są w naszym zasięgu
od punktu startowego do końca zasięgu (pozycja + paliwo). Przy tym sprawdzani zostawiamy
zawsze obecną pozycję (najdalszą), bo tam mogliśmy dotrzeć. Zachłanny wybór najlepszego
w zasięgu rozwiązania gwarantuje nam zawsze dotarcie najdalej, tankując najmniej razy.

### Złożoność

###### Obliczeniowa:
$ O(n \cdot m + k \cdot log(k)) $, gdzie: <br>
$ k $ - liczba rozłącznych plam - złożoność wynika z uzywania kolejki priorytetowej i sortowania indeksów, <br>
$ n $ - szerokość tablicy wejściowej, <br>
$ m $ - wysokość tablicy wejściowej, <br>
###### Pamięciowa:
$ O(k) $ - tablicę otrzymaną na wejściu uznaję za daną i nie wliczam do złożoności.

### Implementacja

```python
class MaxPriorityQueue:
    def __init__(self):
        self._heap = []

    def __len__(self):
        return len(self._heap)

    def insert(self, priority: int, val: object):
        if not isinstance(priority, int):
            raise TypeError(f"priority must be 'int', not {str(type(priority))[7:-1]}")
        # Add a value as the last node of our Complete Binary Tree
        self._heap.append((priority, val))
        # Fix heap in order to satisfy a max-heap property
        self._heapify_up(len(self) - 1)

    # Removes the first value in a priority queue (of a greatest priority)
    def poll(self):
        if not self:
            raise IndexError(f'poll from an empty {self.__class__.__name__}')
        # Store a value to be returned
        removed = self._heap[0]
        # Place the last leaf in the root position
        last = self._heap.pop()
        if len(self) > 0:
            self._heap[0] = last
            # Fix a heap in order to stisfy a max-heap property
            self._heapify_down(0, len(self))
        return removed

    def get_first(self):
        return self._heap[0][1] if self._heap else None

    @staticmethod
    def _parent_idx(curr_idx):
        return (curr_idx - 1) // 2

    @staticmethod
    def _left_child_idx(curr_idx):
        return curr_idx * 2 + 1

    @staticmethod
    def _right_child_idx(curr_idx):
        return curr_idx * 2 + 2

    def _swap(self, i, j):
        self._heap[i], self._heap[j] = self._heap[j], self._heap[i]

    def _heapify_up(self, curr_idx, end_idx=0):
        while curr_idx > end_idx:
            parent_idx = self._parent_idx(curr_idx)
            if self._heap[curr_idx][0] > self._heap[parent_idx][0]:
                self._swap(curr_idx, parent_idx)
            curr_idx = parent_idx

    def _heapify_down(self, curr_idx, end_idx):
        while True:
            l = self._left_child_idx(curr_idx)
            r = self._right_child_idx(curr_idx)
            largest_idx = curr_idx

            if l < end_idx:
                if self._heap[l][0] > self._heap[curr_idx][0]:
                    largest_idx = l
                if r < end_idx and self._heap[r][0] > self._heap[largest_idx][0]:
                    largest_idx = r

            if largest_idx != curr_idx:
                self._swap(curr_idx, largest_idx)
                curr_idx = largest_idx
            else:
                break


def tank(T, j):
    n = len(T)
    m = len(T[0])
    total = 0

    def dfs(i, j):
        nonlocal total
        total += T[i][j]
        T[i][j] = 0  # Remove a fuel from this field
        if i > 0 and T[i - 1][j]:
            dfs(i - 1, j)
        if i < n - 1 and T[i + 1][j]:
            dfs(i + 1, j)
        if j > 0 and T[i][j - 1]:
            dfs(i, j - 1)
        if j < m - 1 and T[i][j + 1]:
            dfs(i, j + 1)

    dfs(0, j)
    return total


def map_array(T):
    m = len(T[0])
    result = []
    for j in range(m):
        if T[0][j]:
            result.append((j, tank(T, j)))
    return result


def plan(T):
    result = [0]
    dist = len(T[0]) - 1

    # Map array
    S = map_array(T)
    fuel = S[0][1]
    pos = 0
    n = len(S)
    i = 1

    pq = MaxPriorityQueue()

    # Add stations which we can reach from the source
    while i < n and S[i][0] <= pos + fuel:
        pq.insert(S[i][1], S[i][0])
        i += 1

    # In a loop take the best station in range [0, pos + fuel]
    while pos + fuel < dist:
        amount, idx = pq.poll()
        # Update fuel amount. If an index 'idx' is larger than the
        # current position, we have to subtract amount of fuel we will
        # use to rach the station which is in our range. Otherwise, if
        # we "go back" to refuel, we don't lose any amount of fuel so
        # we only add amount refueled on this station.
        fuel = fuel - max(idx - pos, 0) + amount
        pos = max(pos, idx)
        result.append(idx)

        # Add stations which we can reach from the current station
        while i < n and S[i][0] <= pos + fuel:
            pq.insert(S[i][1], S[i][0])
            i += 1

    return sorted(result)
```
