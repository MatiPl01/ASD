### Opis algorytmu

Zdecydowałem się na wybór poniższego algorytmu, ponieważ jest on łatwy w zrozumieniu oraz szybki. Jego złożoność obliczeniowa wynosi $ O(|u| + |v| + |w|) $, gdzie $ |u| $, $ |v| $, $ |w| $ - długości stringów, odpowiednio: $ u $, $ v $ oraz $ w $. Algorytm polega na tym, że przechodzimy liniowo przez kolejne znaki stringów $ u $ oraz $ v $ i zliczamy w tablicy liczników wystąpienia każdego ze znaków. Następnie przechodzimy w pętli po znakach ze stringu $ w $ i sprawdzamy, czy dany znak istniał w ciągu tekstowym $ u $ lub $ v $, a więc, czy licznik dla danego znaku jest dodatni. Jeżeli nie, to oznacza, że nie można utworzyć wyrazu $ w $, przy pomocy liter z $ u $ i $ v $. W przeciwnym razie, zmniejszamy licznik dla danego znaku o 1, bo wykorzystujemy ten znak do utworzenia wyrazu $ w $, a następnie przechodzimy do kolejnego znaku z wyrazu $ w $.

### Implementacja algorytmu

```python
def possible(u, v, w, min_char='a', max_char='z'):
    min_ord = ord(min_char)
    # Store counts of each letter of both 'u' and 'v' strings
    counts = [0] * (ord(max_char) - min_ord + 1)
    # Loop over both 'u' and 'v' strings and count their characters
    for char in u: counts[ord(char) - min_ord] += 1
    for char in v: counts[ord(char) - min_ord] += 1
    # Loop over chars in 'w' string and decrement appropriate counters
    # checking in each loop whether a counter is positive.
    # If a value of a counter is 0, return False as we have
    # no remaining characters necessary to create a 'w' string.
    for char in w:
        idx = ord(char) - min_ord
        counts[idx] -= 1
        if counts[idx] < 0:
            return False
    # Return True if it is possible to create 'w' string from 'u' and 'v'
    return True
```
